```python
def build_central_ad_dict(df_age_gender, df_platform, id_col="ad_id"):
    central_dict = {}

    ids_age_gender = set(df_age_gender[id_col].unique())
    ids_platform = set(df_platform[id_col].unique())

    all_ids = ids_age_gender.union(ids_platform)

    for ad_id in all_ids:
        central_dict[ad_id] = {
            "age_gender": None,
            "platform": None,
            "flags": {
                "has_age_gender": ad_id in ids_age_gender,
                "has_platform": ad_id in ids_platform,
                "in_both": ad_id in ids_age_gender and ad_id in ids_platform
            }
        }

    #  age_gender
    for _, row in df_age_gender.iterrows():
        ad_id = row[id_col]
        central_dict[ad_id]["age_gender"] = row.to_dict()

    #  platform
    for _, row in df_platform.iterrows():
        ad_id = row[id_col]
        central_dict[ad_id]["platform"] = row.to_dict()

    return central_dict



central_ads = build_central_ad_dict(df_age_gender, df_publisher_platform, "ad_id")

#
print(f"סה\"כ מודעות במילון: {len(central_ads)}")
sample_id = next(iter(central_ads))
print(sample_id, central_ads[sample_id]["flags"])
```

```python
def central_dict_stats(central_dict):
    total_ads = len(central_dict)
    in_both = sum(1 for v in central_dict.values() if v["flags"]["in_both"])
    only_age_gender = sum(1 for v in central_dict.values() if v["flags"]["has_age_gender"] and not v["flags"]["has_platform"])
    only_platform = sum(1 for v in central_dict.values() if v["flags"]["has_platform"] and not v["flags"]["has_age_gender"])

    stats = {
        "total_ads": total_ads,
        "in_both": in_both,
        "only_age_gender": only_age_gender,
        "only_platform": only_platform,
        "pct_in_both": in_both / total_ads * 100,
        "pct_only_age_gender": only_age_gender / total_ads * 100,
        "pct_only_platform": only_platform / total_ads * 100
    }
    return stats

# חישוב סטטיסטיקות
stats_flags = central_dict_stats(central_ads)

# הצגה
print("📊 סטטיסטיקות דגלים במילון המרכזי:")
for k, v in stats_flags.items():
    print(f"{k}: {v}")
```

```python
import json
import ast
import pandas as pd

# פונקציה קיימת שלך
def safe_parse_actions(raw_actions):
    if pd.isna(raw_actions):
        return []
    if isinstance(raw_actions, list):
        return raw_actions
    if isinstance(raw_actions, str):
        s = raw_actions.strip()
        if not s or s in ("[]", "{}", "nan", "None", "null"):
            return []
        try:
            return json.loads(s)
        except json.JSONDecodeError:
            pass
        try:
            return json.loads(s.replace("'", '"'))
        except json.JSONDecodeError:
            pass
        try:
            parsed = ast.literal_eval(s)
            if isinstance(parsed, list):
                return parsed
            return []
        except Exception:
            return []
    return []

```

```python



# מיפוי הפונקציה
def map_action_type(action_type):
    for unified, aliases in ACTION_MAPPING.items():
        if action_type in aliases:
            return unified
    return f"UNMAPPED::{action_type}"  # אם לא במיפוי – נסמן

# מיפוי כל רשימת הפעולות ברשומה
def map_actions_list(raw_actions):
    parsed = safe_parse_actions(raw_actions)
    mapped = []
    for act in parsed:
        atype = act.get("action_type")
        if atype:
            mapped.append({
                **act,
                "action_type": map_action_type(atype)
            })
    return mapped

# החלת המיפוי על כל DataFrame
df_age_gender["actions_mapped"] = df_age_gender["actions"].apply(map_actions_list)
df_publisher_platform["actions_mapped"] = df_publisher_platform["actions"].apply(map_actions_list)

# --- בדיקת כיסוי ---
all_action_types = set()
for df in [df_age_gender, df_publisher_platform]:
    for raw_actions in df["actions"]:
        for act in safe_parse_actions(raw_actions):
            if "action_type" in act:
                all_action_types.add(act["action_type"])

mapped_values = {alias for aliases in ACTION_MAPPING.values() for alias in aliases}
unmapped_actions = sorted(all_action_types - mapped_values)

print(f"✅ סה\"כ פעולות מקוריות: {len(all_action_types)}")
print(f"⚠️ פעולות שלא במיפוי: {len(unmapped_actions)}")
print(unmapped_actions[:20])  # הדפסה לדוגמה, אפשר להסיר
```

```python
# בניית lookup מהמיפוי
ACTION_LOOKUP = {}
for unified, variants in ACTION_MAPPING.items():
    for v in variants:
        ACTION_LOOKUP[v.strip().lower()] = unified

def add_actions_to_central_ads(central_ads, df, id_col, actions_col, category_cols, target_prefix):
    """
    מוסיף מידע על פעולות ממופות ל-central_ads
    target_prefix - שם הבסיס (למשל "age_gender" => נשמר כ-age_gender_actions)
    """
    target_key = f"{target_prefix}_actions"

    for _, row in df.iterrows():
        actions_list = safe_parse_actions(row[actions_col])
        if not actions_list:
            continue

        ad_id = row[id_col]
        if target_key not in central_ads[ad_id]:
            central_ads[ad_id][target_key] = []

        categories = {col: row[col] for col in category_cols if col in df.columns}
        actions_dict = {}

        for act in actions_list:
            raw_action_type = str(act.get("action_type", "")).strip()
            if not raw_action_type:
                continue

            # המרה לשם אחיד
            action_type = ACTION_LOOKUP.get(raw_action_type.lower(), raw_action_type)

            actions_dict.setdefault(action_type, {})
            for key, val in act.items():
                if key == "action_type":
                    continue
                try:
                    val_num = float(val)
                except (ValueError, TypeError):
                    val_num = 0.0
                actions_dict[action_type][key] = actions_dict[action_type].get(key, 0.0) + val_num

        if actions_dict:
            central_ads[ad_id][target_key].append({
                "categories": categories,
                "actions": actions_dict
            })

# שימוש לדאטה של age_gender
add_actions_to_central_ads(
    central_ads,
    df_age_gender,
    id_col="ad_id",
    actions_col="actions",
    category_cols=["age", "gender", "date_start", "date_stop"],
    target_prefix="age_gender"
)

# שימוש לדאטה של platform
add_actions_to_central_ads(
    central_ads,
    df_publisher_platform,
    id_col="ad_id",
    actions_col="actions",
    category_cols=["publisher_platform", "objective", "date_start", "date_stop"],
    target_prefix="platform"
)

# בדיקה
some_id = next(iter(central_ads))
print(some_id)
print(central_ads[some_id].get("age_gender_actions", []))
print(central_ads[some_id].get("platform_actions", []))
```

```python
# פונקציה למיפוי action_values לפי מילון
def map_action_values(actions_list, mapping):
    """
    מקבל רשימת פעולות גולמית ומחזיר רשימה של פעולות ממופות
    לפי מילון המיפוי שניתן.
    """
    mapped = []
    for action in actions_list:
        action_type = action.get("action_type")
        if not action_type:
            continue
        # חיפוש במיפוי
        for key, aliases in mapping.items():
            if action_type in aliases:
                mapped.append(key)
                break
        else:
            # אם לא נמצא במיפוי - שומרים כמו שהוא
            mapped.append(action_type)
    return mapped


# --- יישום על DataFrame ---
df_age_gender["mapped_action_values"] = df_age_gender["action_values"].apply(
    lambda x: map_action_values(safe_parse_actions(x), ACTION_VALUES_MAPPING)
)

df_publisher_platform["mapped_action_values"] = df_publisher_platform["action_values"].apply(
    lambda x: map_action_values(safe_parse_actions(x), ACTION_VALUES_MAPPING)
)


# --- בדיקת כיסוי ---
all_action_types = set()
for df in [df_age_gender, df_publisher_platform]:
    for raw_actions in df["action_values"]:
        for act in safe_parse_actions(raw_actions):
            if "action_type" in act:
                all_action_types.add(act["action_type"])

# אוסף כל הערכים שמופיעים במילון המיפוי
mapped_values = {alias for aliases in ACTION_VALUES_MAPPING.values() for alias in aliases}

unmapped_actions = sorted(all_action_types - mapped_values)

print(f"✅ סה\"כ פעולות מקוריות: {len(all_action_types)}")
print(f"⚠️ פעולות שלא במיפוי: {len(unmapped_actions)}")
if unmapped_actions:
    print("פעולות לא ממופות לדוגמה:", unmapped_actions[:20])
else:
    print("✅ כל הפעולות ממופות בהצלחה!")
```

```python
def dict_or_empty(x):
    return x if isinstance(x, dict) else {}

def add_core_row(rows, rec, scope, default_ad_id=None):
    meta = dict_or_empty(rec.get(scope))
    plat = dict_or_empty(rec.get("platform"))
    ag   = dict_or_empty(rec.get("age_gender"))

    ad_id = meta.get("ad_id") or plat.get("ad_id") or ag.get("ad_id") or default_ad_id
    ds = meta.get("date_start")
    de = meta.get("date_stop")

    rows.append({
        "ad_id": ad_id,
        "date_start": ds,
        "date_stop": de,
        "objective": meta.get("objective"),
        "age": meta.get("age") if scope == "age_gender" else None,
        "gender": meta.get("gender") if scope == "age_gender" else None,
        "impressions": float(meta.get("impressions") or 0),
        "clicks": float(meta.get("clicks") or 0),
        "spend": float(meta.get("spend") or 0),
        "frequency": float(meta.get("frequency") or 0),
    })

def base_keys_from(rec, scope, default_ad_id=None):
    meta = dict_or_empty(rec.get(scope))
    plat = dict_or_empty(rec.get("platform"))
    ag   = dict_or_empty(rec.get("age_gender"))

    ad_id = meta.get("ad_id") or plat.get("ad_id") or ag.get("ad_id") or default_ad_id
    ds = meta.get("date_start")
    de = meta.get("date_stop")

    return {
        "ad_id": ad_id,
        "date_start": ds,
        "date_stop": de,
        "objective": meta.get("objective"),
        "age": meta.get("age") if scope == "age_gender" else None,
        "gender": meta.get("gender") if scope == "age_gender" else None,
    }

def build_core_table(master_dict):
    rows = []
    for ad_id_key, rec in master_dict.items():
        if not isinstance(rec, dict):
            continue
        add_core_row(rows, rec, "platform", default_ad_id=ad_id_key)
        add_core_row(rows, rec, "age_gender", default_ad_id=ad_id_key)

    df_core = pd.DataFrame(rows)
    for c in ["impressions","clicks","spend","frequency"]:
        df_core[c] = pd.to_numeric(df_core[c], errors="coerce").fillna(0.0)

    key_cols = ["ad_id","date_start","date_stop","objective","age","gender"]
    df_core = df_core.groupby(key_cols, dropna=False, as_index=False)[
        ["impressions","clicks","spend","frequency"]
    ].sum()
    return df_core

# פונקציה שמוודאת שכל העמודות קיימות
def ensure_columns(df, cols):
    for c in cols:
        if c not in df.columns:
            df[c] = pd.NA
    return df
def emit_event_rows(event_list, base_keys, stat_name):
    """
    מקבלת:
      event_list - רשימת מילונים מה־JSON (למשל [{"action_type":"link_click","value": 123}, ...])
      base_keys  - מילון עם מפתחות הבסיס (ad_id, date_start, date_stop, objective, age, gender)
      stat_name  - שם הסטטיסטיקה (actions, action_values וכו') — נשמר בעמודה 'stat'

    מחזירה:
      רשימת מילונים, כל אחד מייצג שורה בטבלת האירועים.
    """
    rows = []
    if not event_list:
        return rows

    for ev in event_list:
        if not isinstance(ev, dict):
            continue
        row = dict(base_keys)  # העתקת המפתחות הבסיסיים
        row["action_type"] = ev.get("action_type")
        row["stat"] = stat_name
        row["amount"] = ev.get("value") or ev.get("amount") or 0
        rows.append(row)

    return rows

def build_event_tables(master_dict):
    buckets = {
        "actions": [],
        "action_values": [],
        "video_play_actions": [],
        "video_thruplay_watched_actions": [],
        "video_avg_time_watched_actions": [],
    }

    for ad_id_key, rec in master_dict.items():
        if not isinstance(rec, dict):
            continue

        # PLATFORM
        bk = base_keys_from(rec, "platform", default_ad_id=ad_id_key)
        buckets["actions"] += emit_event_rows(rec.get("platform_actions"), bk, "actions")
        buckets["action_values"] += emit_event_rows(rec.get("platform_action_values"), bk, "action_values")
        buckets["video_play_actions"] += emit_event_rows(rec.get("platform_video_play_actions"), bk, "video_play_actions")
        buckets["video_thruplay_watched_actions"] += emit_event_rows(rec.get("platform_video_thruplay_watched_actions"), bk, "video_thruplay_watched_actions")
        buckets["video_avg_time_watched_actions"] += emit_event_rows(rec.get("platform_video_avg_time_watched_actions"), bk, "video_avg_time_watched_actions")

        # AGE-GENDER
        bk2 = base_keys_from(rec, "age_gender", default_ad_id=ad_id_key)
        buckets["actions"] += emit_event_rows(rec.get("age_gender_actions"), bk2, "actions")
        buckets["action_values"] += emit_event_rows(rec.get("age_gender_action_values"), bk2, "action_values")
        buckets["video_play_actions"] += emit_event_rows(rec.get("age_gender_video_play_actions"), bk2, "video_play_actions")
        buckets["video_thruplay_watched_actions"] += emit_event_rows(rec.get("age_gender_video_thruplay_watched_actions"), bk2, "video_thruplay_watched_actions")
        buckets["video_avg_time_watched_actions"] += emit_event_rows(rec.get("age_gender_video_avg_time_watched_actions"), bk2, "video_avg_time_watched_actions")

    dfs = {}
    key_cols = ["ad_id","date_start","date_stop","objective","age","gender","action_type","stat"]

    for name, rows in buckets.items():
        df = pd.DataFrame(rows)
        df = ensure_columns(df, key_cols + ["amount"])
        if not df.empty:
            df["amount"] = pd.to_numeric(df["amount"], errors="coerce").fillna(0.0)
            df = df.groupby(key_cols, dropna=False, as_index=False)["amount"].sum()
        dfs[name] = df

    return (dfs["actions"], dfs["action_values"], dfs["video_play_actions"],
            dfs["video_thruplay_watched_actions"], dfs["video_avg_time_watched_actions"])
```

```python
# ---- 1) נרמול מפתחות בכל הטבלאות ----
import pandas as pd

def normalize_keys(df):
    if df is None or df.empty:
        return df
    out = df.copy()

    # מימדים קטגוריים -> pandas StringDtype, מנקים ערכים ריקים/“None”
    for c in ["objective", "age", "gender"]:
        if c in out.columns:
            out[c] = out[c].astype("string")
            out.loc[out[c].isin(["", "None", "nan"]), c] = pd.NA

    # תאריכים -> parse -> strftime -> StringDtype (אחיד "YYYY-MM-DD")
    for c in ["date_start", "date_stop"]:
        if c in out.columns:
            dt = pd.to_datetime(out[c], errors="coerce")
            out[c] = dt.dt.strftime("%Y-%m-%d").astype("string")
            out.loc[dt.isna(), c] = pd.NA  # מחזירים NA במקום NaT/NaN
    return out

# נרמל core + כל טבלאות האירועים
df_core_n = normalize_keys(df_core)

df_actions_n       = normalize_keys(df_actions)
df_action_values_n = normalize_keys(df_action_values)
df_vplay_n         = normalize_keys(df_video_play_actions)
df_vthru_n         = normalize_keys(df_video_thruplay_watched_actions)
df_vavg_n          = normalize_keys(df_video_avg_time_watched_actions)
```

```python
# ---- 2) בונים קרטזית עם period-pairs אמיתיים (מהגרסה שלך) ----
from itertools import product

SENTINEL = "__NA__"

def _collect_period_pairs_for_ad(ad_id, sources):
    pairs = set()
    for df in sources:
        if df is None or df.empty:
            continue
        sub = df[df["ad_id"] == ad_id]
        if {"date_start","date_stop"}.issubset(sub.columns):
            # אחרי normalize, שניהם כבר string/NA
            for ds, de in zip(sub["date_start"], sub["date_stop"]):
                ds = None if pd.isna(ds) else ds
                de = None if pd.isna(de) else de
                pairs.add((ds, de))
    return list(pairs)  # בלי sorted כדי להימנע מ-TypeError

def _values_by_dim_for_ad(ad_id, dims, sources):
    by_dim = {d: set() for d in dims}
    for df in sources:
        if df is None or df.empty:
            continue
        sub = df[df["ad_id"] == ad_id]
        for d in dims:
            if d in sub.columns:
                by_dim[d].update(sub[d].dropna().unique().tolist())
    return {d: (sorted(list(v)) if v else []) for d, v in by_dim.items()}

def build_cartesian_dates_as_pairs(df_core, event_tables=None, dims=None):
    if dims is None:
        dims = ["objective","age","gender"]
    all_sources = [df_core] + ([t for t in event_tables if t is not None] if event_tables else [])
    rows = []
    for ad_id in df_core["ad_id"].dropna().unique():
        # זוגות זמן אמיתיים אחרי normalize (strings אחידים)
        period_pairs = _collect_period_pairs_for_ad(ad_id, all_sources)
        if not period_pairs:
            period_pairs = [(pd.NA, pd.NA)]

        space = _values_by_dim_for_ad(ad_id, dims, all_sources)
        lists = {d: (space[d] if space[d] else [pd.NA]) for d in dims}

        for ds, de in period_pairs:
            for combo in product(*(lists[d] for d in dims)):
                rec = {"ad_id": ad_id, "date_start": ds, "date_stop": de}
                rec.update({d: v for d, v in zip(dims, combo)})
                rows.append(rec)
    return pd.DataFrame(rows, columns=["ad_id","date_start","date_stop"] + dims)

```

```python
# ---- 3) Anti-join לזיהוי חסרים (כמו שהיה, אבל נוודא keys אחידים) ----
def _ensure_keys(df, keys):
    out = df.copy()
    for k in keys:
        if k not in out.columns:
            out[k] = pd.NA
    return out

def _prep_for_join(df, keys):
    df2 = _ensure_keys(df, keys)[keys].drop_duplicates().copy()
    # כל המפתחות כבר string/NA אחרי normalize; מחליף NA בסנטינל רק בשביל השוואה
    for k in keys:
        df2[k] = df2[k].astype("string").where(df2[k].notna(), SENTINEL)
    return df2


def find_missing_flexible_per_ad(cart, df_event, dims):
    """
    Anti-join גמיש: NA בצד האירוע = wildcard.
    עובד פר-ad_id, מטפל בעמודות כפולות ובערכים לא-סקאלריים (רשימות/Series).
    """
    keys = ["ad_id"] + dims

    # 1) עותקים + הסרת עמודות כפולות
    cart2 = cart.copy()
    ev    = df_event.copy()
    cart2 = cart2.loc[:, ~cart2.columns.duplicated()]
    ev    = ev.loc[:, ~ev.columns.duplicated()]

    # 2) לוודא שכל עמודות המפתח קיימות
    for k in keys:
        if k not in cart2.columns:
            cart2[k] = pd.NA
        if k not in ev.columns:
            ev[k] = pd.NA

    # 3) לעבוד רק עם עמודות המפתח ולנקות כפילויות שורות
    cart2 = cart2[keys].drop_duplicates().copy()
    ev2   = ev[keys].drop_duplicates().copy()

    missing_parts = []

    for ad_id, cart_g in cart2.groupby("ad_id", dropna=False):
        ev_g = ev2[ev2["ad_id"] == ad_id]
        if ev_g.empty:
            # אין בכלל אירועים למודעה → הכל חסר
            missing_parts.append(cart_g)
            continue

        # מסכת כיסוי לשורות הקרטזית של אותו ad_id
        covered = np.zeros(len(cart_g), dtype=bool)

        # נאחסן מראש את ערכי העמודות של הקרטזית כ-numpy
        cart_cols = {d: cart_g[d].values for d in dims}

        for _, rule in ev_g.iterrows():
            m = np.ones(len(cart_g), dtype=bool)

            for d in dims:
                # קבלת ערך/ים של הממד d מה"חוק"
                val = rule[d]

                # אם זה רשימתי/Series → נעשה flatten לערכים ייחודיים לא-NA
                if isinstance(val, (pd.Series, np.ndarray, list, tuple)):
                    allowed = pd.Series(val).dropna().unique()
                else:
                    allowed = [] if pd.isna(val) else [val]

                if len(allowed) == 0:
                    # wildcard - לא מצמצם את m
                    continue

                # התאמה: הערך בקרטזית חייב להיות אחד מהערכים המותרים
                m &= np.isin(cart_cols[d], allowed)

                # אופטימיזציה קטנה: אם כבר ריק, אין טעם לבדוק שדות נוספים
                if not m.any():
                    break

            covered |= m
            if covered.all():
                break

        missing_parts.append(cart_g[~covered])

    return pd.concat(missing_parts, ignore_index=True) if missing_parts else cart2.iloc[0:0]



```

```python
import pandas as pd
import numpy as np

# ---------------------------
# Helpers: weights
# ---------------------------
def _safe_norm(w):
    w = np.asarray(w, dtype=float)
    s = w.sum()
    if s <= 0 or not np.isfinite(s):
        # אחיד
        return np.ones_like(w) / len(w)
    return w / s

def _weights_age_gender(cart_g, present_g=None, fallback_g=None, weight_col="impressions"):
    """
    מחזיר וקטור משקולות לכל שורה ב-cart_g לפי:
    1) אם יש present_g עם weight_col -> משתמש בהתפלגות בפועל (לפי age,gender).
    2) אחרת אם יש fallback_g (למשל aggregation עוקב) -> משתמש בה.
    3) אחרת -> אחיד.
    """
    # cart_g: השורות שנרצה לחלק ביניהן (מימדים: age,gender)
    key = ["age","gender"]

    # 1) ניסיון: לפי present_g
    if present_g is not None and not present_g.empty and weight_col in present_g:
        pg = present_g.groupby(key, dropna=False, as_index=False)[weight_col].sum()
        merged = cart_g[key].merge(pg, on=key, how="left")
        w = merged[weight_col].fillna(0).values
        if w.sum() > 0:
            return _safe_norm(w)

    # 2) ניסיון: לפי fallback
    if fallback_g is not None and not fallback_g.empty and weight_col in fallback_g:
        fb = fallback_g.groupby(key, dropna=False, as_index=False)[weight_col].sum()
        merged = cart_g[key].merge(fb, on=key, how="left")
        w = merged[weight_col].fillna(0).values
        if w.sum() > 0:
            return _safe_norm(w)

    # 3) אחיד
    return np.ones(len(cart_g)) / max(len(cart_g), 1)

# ---------------------------
# Delivery estimation (core)
# ---------------------------
def estimate_core_metrics(df_core, cart,
                          metrics=("impressions","clicks","spend"),
                          dims=("objective","age","gender","date_start","date_stop")):
    """
    ממלא חסרים ב-core כך שסכומי המטריקות פר (ad_id, period, objective) יתאימו ליעד הפלטפורמה.
    - cart: הקרטזית המלאה (עם age,gender,date_start,date_stop,objective)
    מחזיר df_core_est באותו גרעין + המטריקות.
    """
    keys = ["ad_id"] + list(dims)
    # חיתוך על העמודות הרלוונטיות
    core_cols = list(set(keys) | set(metrics))
    core = df_core[core_cols].copy()

    # מפרידים רשומות פלטפורמה (age,gender = NA) לעומת age_gender
    is_plat = core["age"].isna() & core["gender"].isna()
    core_plat = core[is_plat].copy()
    core_ag   = core[~is_plat].copy()

    # נחזיר טבלת “מלאה” על בסיס cart
    out = cart[keys].drop_duplicates().copy()

    # נאתר missing מול מה שיש ב-core_ag
    present_keys = core_ag[keys].drop_duplicates()
    out = out.merge(present_keys.assign(_present=1), on=keys, how="left")
    missing_mask = out["_present"].isna()
    need_fill = out[missing_mask].drop(columns="_present")
    out = out.drop(columns="_present")

    # נחבר חזרה את מה שכבר קיים
    out = out.merge(core_ag, on=keys, how="left")

    # נריץ פר-קבוצה: (ad_id, date_start, date_stop, objective)
    grp_keys = ["ad_id","date_start","date_stop","objective"]

    # fallback גלובלי: התפלגות היסטורית למודעה על פני כל התקופות (age,gender) לפי impressions
    hist = core_ag.groupby(["ad_id","age","gender"], dropna=False, as_index=False)["impressions"].sum()

    # נבנה אינדקס למילוי מהיר
    need_fill = need_fill.merge(cart[["ad_id","date_start","date_stop","objective","age","gender"]].drop_duplicates(),
                                on=["ad_id","date_start","date_stop","objective","age","gender"], how="left")

    for gvals, cart_g in need_fill.groupby(grp_keys, dropna=False):
        ad_id, ds, de, obj = gvals

        # ה-Target מהרשומת פלטפורמה (אם קיימת)
        plat_row = core_plat[(core_plat["ad_id"]==ad_id) &
                             (core_plat["date_start"]==ds) &
                             (core_plat["date_stop"]==de) &
                             (core_plat["objective"]==obj)]
        # השורות שכבר קיימות לגיל/מין באותה קבוצה
        present_g = core_ag[(core_ag["ad_id"]==ad_id) &
                            (core_ag["date_start"]==ds) &
                            (core_ag["date_stop"]==de) &
                            (core_ag["objective"]==obj)]

        # fallback היסטורי למודעה
        hist_g = hist[hist["ad_id"]==ad_id].rename(columns={"impressions":"weight"})

        # החתך בקרטזית שצריך מילוי (age,gender)
        to_fill_idx = need_fill.index.isin(cart_g.index)
        cart_slice = cart_g.copy()

        # משקולות לפי impressions (עדיף, כי הכי יציב)
        w = _weights_age_gender(cart_slice,
                                present_g.rename(columns={"impressions":"impressions"}),
                                hist_g.rename(columns={"weight":"impressions"}),
                                weight_col="impressions")

        # לכל מטריקה: מחשבים Target, Observed, Residual ומחלקים לפי w
        for m in metrics:
            target = float(plat_row[m].sum()) if (not plat_row.empty and m in plat_row) else np.nan
            observed = float(present_g[m].sum()) if (m in present_g) else 0.0

            # אם אין Target – נשתמש בסכום שכבר קיים (כלומר לא נכפה איזון)
            if not np.isfinite(target):
                continue

            residual = target - observed
            if abs(residual) < 1e-9:
                # אין מה לחלק
                continue

            alloc = residual * w
            # הזרקה לטבלת out
            mask_insert = (
                (out["ad_id"]==ad_id) &
                (out["date_start"]==ds) &
                (out["date_stop"]==de) &
                (out["objective"]==obj) &
                out["age"].isin(cart_slice["age"]) &
                out["gender"].isin(cart_slice["gender"])
            )
            # אם העמודה לא קיימת עדיין (NaN), נתחיל מ-0
            if m not in out.columns:
                out[m] = 0.0
            out.loc[mask_insert & out[m].isna(), m] = 0.0

            # התאמה לפי סדר השורות
            # נבנה אינדקסים תואמים:
            targ_idx = out[mask_insert].sort_values(["age","gender"]).index
            add_vals = pd.Series(alloc, index=out[mask_insert].sort_values(["age","gender"]).index)
            out.loc[targ_idx, m] = out.loc[targ_idx, m].fillna(0.0) + add_vals

    # אחר טיפול: אם יש עדיין NaN במטריקות – לא נגעו בהם, נמלא 0
    for m in metrics:
        if m in out.columns:
            out[m] = pd.to_numeric(out[m], errors="coerce").fillna(0.0)

    return out

# ---------------------------
# Events estimation (actions & co.)
# ---------------------------
def estimate_event_table(df_event, cart, df_core_est,
                         dims=("objective","age","gender","date_start","date_stop"),
                         amount_col="amount", id_cols=("ad_id","action_type","stat"),
                         weight_metric_for_split="impressions"):
    """
    ממלא חסרים בטבלת אירועים כך שסכומי amount פר (ad_id, period, objective, action_type, stat)
    יתאימו ליעד הפלטפורמה (age,gender=NA) אם קיים; אחרת נשמר סכום קיים.
    פיזור לפי משקולות מה-core_est (לפי weight_metric_for_split, ברירת-מחדל impressions).
    """
    keys = ["ad_id"] + list(dims)
    ev_cols = list(set(keys) | set(id_cols) | {amount_col})
    ev = df_event[ev_cols].copy()

    # מפרידים פלטפורמה לעומת age_gender
    is_plat = ev["age"].isna() & ev["gender"].isna()
    ev_plat = ev[is_plat].copy()
    ev_ag   = ev[~is_plat].copy()

    # cart מורחב עם action_type/stat לכל מה שקיים ב-ev
    act_stat = ev[list(id_cols)].drop_duplicates()
    base = cart[keys].drop_duplicates()
    full_space = base.merge(act_stat, on=["ad_id"], how="inner")

    # אילו חסרים?
    present_keys = ev_ag[keys + list(id_cols)].drop_duplicates()
    to_fill = full_space.merge(present_keys.assign(_p=1),
                               on=keys+list(id_cols), how="left")
    to_fill = to_fill[to_fill["_p"].isna()].drop(columns="_p")

    out = full_space.merge(ev_ag, on=keys+list(id_cols), how="left")

    grp_keys = ["ad_id","date_start","date_stop","objective","action_type","stat"]

    # משקולות לפיזור מתוך core_est
    core_w = df_core_est[keys + [weight_metric_for_split]].copy()

    for gvals, cart_g in to_fill.groupby(grp_keys, dropna=False):
        ad_id, ds, de, obj, act, stat = gvals

        # Target מהפלטפורמה (age,gender=NA) לאותה קבוצה
        plat_row = ev_plat[(ev_plat["ad_id"]==ad_id) &
                           (ev_plat["date_start"]==ds) &
                           (ev_plat["date_stop"]==de) &
                           (ev_plat["objective"]==obj) &
                           (ev_plat["action_type"]==act) &
                           (ev_plat["stat"]==stat)]
        present_g = ev_ag[(ev_ag["ad_id"]==ad_id) &
                          (ev_ag["date_start"]==ds) &
                          (ev_ag["date_stop"]==de) &
                          (ev_ag["objective"]==obj) &
                          (ev_ag["action_type"]==act) &
                          (ev_ag["stat"]==stat)]

        # אם אין Target – אין למה לאזן (נשאיר כמו שיש)
        target = float(plat_row[amount_col].sum()) if (not plat_row.empty) else np.nan
        if not np.isfinite(target):
            continue

        observed = float(present_g[amount_col].sum()) if (amount_col in present_g) else 0.0
        residual = target - observed
        if abs(residual) < 1e-9:
            continue

        # משקולות לפיזור לפי core_est (impressions ברירת-מחדל)
        core_slice = core_w[(core_w["ad_id"]==ad_id) &
                            (core_w["date_start"]==ds) &
                            (core_w["date_stop"]==de) &
                            (core_w["objective"]==obj)]

        w = _weights_age_gender(cart_g, present_g=None, fallback_g=core_slice,
                                weight_col=weight_metric_for_split)

        alloc = residual * w

        mask_insert = (
            (out["ad_id"]==ad_id) &
            (out["date_start"]==ds) &
            (out["date_stop"]==de) &
            (out["objective"]==obj) &
            (out["action_type"]==act) &
            (out["stat"]==stat) &
            out["age"].isin(cart_g["age"]) &
            out["gender"].isin(cart_g["gender"])
        )

        if amount_col not in out.columns:
            out[amount_col] = 0.0
        out.loc[mask_insert & out[amount_col].isna(), amount_col] = 0.0

        targ_idx = out[mask_insert].sort_values(["age","gender"]).index
        add_vals = pd.Series(alloc, index=targ_idx)
        out.loc[targ_idx, amount_col] = out.loc[targ_idx, amount_col].fillna(0.0) + add_vals.values

    # ניקוי NaN
    out[amount_col] = pd.to_numeric(out[amount_col], errors="coerce").fillna(0.0)
    return out
```

```python
def estimate_event_table(
    df_event, cart, df_core_est,
    dims=("objective","age","gender","date_start","date_stop"),
    amount_col="amount", id_cols=("ad_id","action_type","stat"),
    weight_metric_for_split="impressions"
):
    import pandas as pd
    import numpy as np

    def _uniq(seq):
        out, seen = [], set()
        for x in seq:
            if x not in seen:
                seen.add(x); out.append(x)
        return out

    def _drop_dup_cols(df):
        return df.loc[:, ~df.columns.duplicated()].copy()

    keys    = ["ad_id"] + list(dims)
    id_cols = list(id_cols)

    # עמודות שנצטרך מה-event (בלי כפילויות)
    ev_cols = _uniq(keys + id_cols + [amount_col])
    ev = _drop_dup_cols(df_event[ev_cols].copy())

    # פלטפורמה (NA/NA) מול גיל-מין
    is_plat = ev["age"].isna() & ev["gender"].isna()
    ev_plat = ev[is_plat].copy()
    ev_ag   = ev[~is_plat].copy()

    # cart בסיסי
    base = _drop_dup_cols(cart[_uniq(keys)].drop_duplicates().copy())

    # action_type/stat שקיימים בטבלה (בלי כפילויות)
    act_stat = _drop_dup_cols(ev[_uniq(id_cols)].drop_duplicates())

    # מרחב מלא (ad_id × dims × action_type/stat)
    full_space = _drop_dup_cols(base.merge(act_stat, on=["ad_id"], how="inner"))

    # אילו קומבינציות כבר קיימות בגיל/מין?
    join_cols = _uniq(keys + id_cols)              # מפתחות למיזוג (ללא כפילויות!)
    present_keys = _drop_dup_cols(ev_ag[join_cols].drop_duplicates())

    to_fill = full_space.merge(present_keys.assign(_p=1), on=join_cols, how="left")
    to_fill = to_fill[to_fill["_p"].isna()].drop(columns="_p")

    out = _drop_dup_cols(full_space.merge(ev_ag, on=join_cols, how="left"))

    grp_keys = ["ad_id","date_start","date_stop","objective","action_type","stat"]

    # משקולות לפיזור מה-core_est
    core_w = _drop_dup_cols(df_core_est[_uniq(keys + [weight_metric_for_split])].copy())

    # פונקציית משקולות קיימת אצלך (_weights_age_gender)

    for gvals, cart_g in to_fill.groupby(grp_keys, dropna=False):
        ad_id, ds, de, obj, act, stat = gvals

        plat_row = ev_plat[
            (ev_plat["ad_id"]==ad_id) &
            (ev_plat["date_start"]==ds) & (ev_plat["date_stop"]==de) &
            (ev_plat["objective"]==obj) &
            (ev_plat["action_type"]==act) & (ev_plat["stat"]==stat)
        ]
        present_g = ev_ag[
            (ev_ag["ad_id"]==ad_id) &
            (ev_ag["date_start"]==ds) & (ev_ag["date_stop"]==de) &
            (ev_ag["objective"]==obj) &
            (ev_ag["action_type"]==act) & (ev_ag["stat"]==stat)
        ]

        target = float(plat_row[amount_col].sum()) if not plat_row.empty else np.nan
        if not np.isfinite(target):
            continue

        observed = float(present_g[amount_col].sum()) if (amount_col in present_g) else 0.0
        residual = target - observed
        if abs(residual) < 1e-9:
            continue

        core_slice = core_w[
            (core_w["ad_id"]==ad_id) &
            (core_w["date_start"]==ds) & (core_w["date_stop"]==de) &
            (core_w["objective"]==obj)
        ][["age","gender", weight_metric_for_split]]

        w = _weights_age_gender(
            cart_g[["age","gender"]],
            present_g=None,
            fallback_g=core_slice.rename(columns={weight_metric_for_split:"impressions"}),
            weight_col="impressions",
        )

        mask_insert = (
            (out["ad_id"]==ad_id) &
            (out["date_start"]==ds) & (out["date_stop"]==de) &
            (out["objective"]==obj) &
            (out["action_type"]==act) & (out["stat"]==stat) &
            out["age"].isin(cart_g["age"]) & out["gender"].isin(cart_g["gender"])
        )

        if amount_col not in out.columns:
            out[amount_col] = 0.0
        out.loc[mask_insert & out[amount_col].isna(), amount_col] = 0.0

        targ_idx = out[mask_insert].sort_values(["age","gender"]).index
        out.loc[targ_idx, amount_col] = (
            out.loc[targ_idx, amount_col].fillna(0.0) + pd.Series(w, index=targ_idx) * residual
        )

    out[amount_col] = pd.to_numeric(out[amount_col], errors="coerce").fillna(0.0)
    return _drop_dup_cols(out)
```

```python
# ---- 2) בונים קרטזית עם period-pairs אמיתיים (מהגרסה שלך) ----
from itertools import product

SENTINEL = "__NA__"

def _collect_period_pairs_for_ad(ad_id, sources):
    pairs = set()
    for df in sources:
        if df is None or df.empty:
            continue
        sub = df[df["ad_id"] == ad_id]
        if {"date_start","date_stop"}.issubset(sub.columns):
            # אחרי normalize, שניהם כבר string/NA
            for ds, de in zip(sub["date_start"], sub["date_stop"]):
                ds = None if pd.isna(ds) else ds
                de = None if pd.isna(de) else de
                pairs.add((ds, de))
    return list(pairs)  # בלי sorted כדי להימנע מ-TypeError

def _values_by_dim_for_ad(ad_id, dims, sources):
    by_dim = {d: set() for d in dims}
    for df in sources:
        if df is None or df.empty:
            continue
        sub = df[df["ad_id"] == ad_id]
        for d in dims:
            if d in sub.columns:
                by_dim[d].update(sub[d].dropna().unique().tolist())
    return {d: (sorted(list(v)) if v else []) for d, v in by_dim.items()}

def build_cartesian_dates_as_pairs(df_core, event_tables=None, dims=None):
    if dims is None:
        dims = ["objective","age","gender"]
    all_sources = [df_core] + ([t for t in event_tables if t is not None] if event_tables else [])
    rows = []
    for ad_id in df_core["ad_id"].dropna().unique():
        # זוגות זמן אמיתיים אחרי normalize (strings אחידים)
        period_pairs = _collect_period_pairs_for_ad(ad_id, all_sources)
        if not period_pairs:
            period_pairs = [(pd.NA, pd.NA)]

        space = _values_by_dim_for_ad(ad_id, dims, all_sources)
        lists = {d: (space[d] if space[d] else [pd.NA]) for d in dims}

        for ds, de in period_pairs:
            for combo in product(*(lists[d] for d in dims)):
                rec = {"ad_id": ad_id, "date_start": ds, "date_stop": de}
                rec.update({d: v for d, v in zip(dims, combo)})
                rows.append(rec)
    return pd.DataFrame(rows, columns=["ad_id","date_start","date_stop"] + dims)


# ---- 3) Anti-join לזיהוי חסרים (כמו שהיה, אבל נוודא keys אחידים) ----
def _ensure_keys(df, keys):
    out = df.copy()
    for k in keys:
        if k not in out.columns:
            out[k] = pd.NA
    return out

def _prep_for_join(df, keys):
    df2 = _ensure_keys(df, keys)[keys].drop_duplicates().copy()
    # כל המפתחות כבר string/NA אחרי normalize; מחליף NA בסנטינל רק בשביל השוואה
    for k in keys:
        df2[k] = df2[k].astype("string").where(df2[k].notna(), SENTINEL)
    return df2

def find_missing(cart, df_event, dims=None):
    if dims is None:
        dims = ["objective","age","gender","date_start","date_stop"]
    keys = ["ad_id"] + dims
    cart_j = _prep_for_join(cart, keys)
    pres_j = _prep_for_join(df_event, keys)
    merged = cart_j.merge(pres_j, on=keys, how="left", indicator=True)
    miss = merged[merged["_merge"]=="left_only"].drop(columns="_merge").copy()
    for k in keys:
        miss[k] = miss[k].replace(SENTINEL, pd.NA)
    return miss
```

```python
import itertools
import pandas as pd

def find_missing_combinations(unified_df, cat_cols=['age','gender','publisher_platform','objective']):
    missing_rows = []

    # נוודא שהקטגוריות הן באמת עמודות בטבלה
    cat_cols = [c for c in cat_cols if c in unified_df.columns]

    for ad_id, group in unified_df.groupby('ad_id'):
        # 1. סטים של ערכים קיימים לכל ממד
        dims_values = {col: sorted(group[col].dropna().unique()) for col in cat_cols}

        # אם באחד המימדים אין ערכים (כל NaN) — מדלגים
        if any(len(vals) == 0 for vals in dims_values.values()):
            continue

        # 2. כל הצירופים האפשריים
        all_combos = list(itertools.product(*dims_values.values()))

        # 3. צירופים קיימים בפועל
        existing_combos = set(tuple(row) for row in group[cat_cols].itertuples(index=False, name=None))

        # 4. צירופים שחסרים
        for combo in all_combos:
            if combo not in existing_combos:
                row_dict = {'ad_id': ad_id}
                row_dict.update(dict(zip(cat_cols, combo)))
                missing_rows.append(row_dict)

    missing_df = pd.DataFrame(missing_rows)
    return missing_df

# דוגמת הרצה
missing_combos_df = find_missing_combinations(unified)
print("Missing combinations found:", len(missing_combos_df))
missing_combos_df.head()
```

```python
def estimate_missing_values(unified_df, missing_df, metrics=['impressions','clicks','spend','frequency'], cat_cols=['age','gender','publisher_platform','objective']):
    # נאחד הכל בסוף ל-DataFrame אחד
    est_rows = []

    for ad_id, group in unified_df.groupby('ad_id'):
        # סכום מטריקות בקבוצת המקור
        totals = {m: group[m].sum(skipna=True) for m in metrics if m in group.columns}

        # כל הצירופים האפשריים לפי קטגוריות של אותו ad_id
        dims_values = {col: sorted(group[col].dropna().unique()) for col in cat_cols if col in group.columns}
        if any(len(vals) == 0 for vals in dims_values.values()):
            continue
        import itertools
        all_combos = list(itertools.product(*dims_values.values()))

        # צירופים קיימים
        existing_combos = set(tuple(row) for row in group[cat_cols].itertuples(index=False, name=None))

        # צירופים חסרים
        missing_combos = [c for c in all_combos if c not in existing_combos]
        n_missing = len(missing_combos)
        if n_missing == 0:
            continue

        # אומדן אחיד: נחלק את "היתרה" לכל החסרים באופן שווה
        for combo in missing_combos:
            row_dict = {'ad_id': ad_id}
            row_dict.update(dict(zip(cat_cols, combo)))
            for m in metrics:
                if m in totals:
                    row_dict[m] = totals[m] / (len(existing_combos) + n_missing)
            est_rows.append(row_dict)

    est_df = pd.DataFrame(est_rows)
    return est_df

# דוגמת הרצה
estimated_missing_df = estimate_missing_values(
    unified,
    missing_combos_df,
    metrics=['impressions','clicks','spend','frequency'],
    cat_cols=['age','gender','publisher_platform','objective']
)

print("Estimated missing rows:", estimated_missing_df.shape)
estimated_missing_df.head()
```
